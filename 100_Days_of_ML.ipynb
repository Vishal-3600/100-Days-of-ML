🔹 **1. Creating Arrays**

```python
import numpy as np

# 1D array
a=np.array([1,2,3,4])

#2D array
b=np.array([[1,2,3,],
          [5,6,7]])# in this 2D you must have same no of elements in its lower dimensions

c = np.array([[[1,2],[3,4]],
              [[5,6],[7,8]]])
print(a)
print(b)
print(c)
```

**2. Important Attributes**

```python
print(a.shape)# dimensions(meaning: 4 elements and 1 dimensions)
print(b.shape)#(Meaning 2 rows and 3 columns)
print(c.shape)#(meaning 2 block 2 rows and 2 columns) This is 3D array
'''
Layer 1
1 2
3 4

Layer 2
5 6
7 8

'''
print(a.ndim) #number of axes(dimenisons)
print(b.ndim)
print(c.ndim)
print(a.dtype)# dataType
```

**3. Vectorized Operations (Very Important)**

```python
x = np.array([1,2,3])
y = np.array([4,5,6])

print(x + y)      # element-wise addition
print(x * y)      # element-wise multiplication
print()

#2D arrays
x = np.array([[1,2,3],
              [4,5,6]])

y = np.array([[10,20,30],
              [40,50,60]])


print(x + y)
print(x * y) #NumPy still does element-wise operations, but shapes must be compatible.
print()

#3D arrays
x = np.array([[[1,2],[3,4]],
              [[5,6],[7,8]]])

y = np.array([[[1,1],[1,1]],
              [[2,2],[2,2]]])

print(x + y)
print(x * y) #Still element-wise:(Same block(2), Same row(2), Same column(2))

#If shapes don’t match:
x = np.array([1,2,3])
y = np.array([[1,2,3],
              [4,5,6]])

print(x + y)
#This works because of Broadcasting.

#What is Broadcasting?(lean more about it)

#  NumPy automatically expands smaller array to match bigger one.

#x = [1 2 3]
# this becomes
'''[1 2 3]
    [1 2 3]'''
```

**🔹 4. Dot Product (Core ML Concept)**

Used in:

Linear Regression    ----> Y=Xw

Neural Networks

SVM

Logistic Regression

```python
x = np.array([1,2,3])
y = np.array([4,5,6])

dot = np.dot(x, y)
print(dot) #1×4+2×5+3×6=4+10+18=32
print()

#2D × 2D (Matrix Multiplication)

x = np.array([[1,2],
              [3,4]])

y = np.array([[5,6],
              [7,8]])

print(np.dot(x, y))#(Am×n​)⋅(Bn×p​)=(m×p)((2×2) · (2×2) → (2×2))
print()


#2D × 1D
x = np.array([[1,2,3],
              [4,5,6]])   # shape (2,3)

y = np.array([1,1,1])     # shape (3,)
print(np.dot(x, y))
```

**5. Matrix Multiplication**

```python
#(2D × 2D)
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

print(np.matmul(A, B))

#3D × 3D (Batch Matrix Multiplication)
A = np.ones((2, 2, 3))   # shape (batch=2, 2, 3)
B = np.ones((2, 3, 4))   # shape (batch=2, 3, 4)

C = np.matmul(A, B)
print(C.shape)
print()
# Broadcasting in matmul
A = np.ones((2, 2, 3))   # (2,2,3)
B = np.ones((3, 4))      # (3,4) it will be treated as (1,3,4)

C = np.matmul(A, B) #(2,2,3) X (2,3,4) =(2,2,4)
print(C.shape)

# Rules
#Batch → Rows → Columns
#(last-2-dim) × (last-2-dim)
#All other dimensions are treated as batches.
```
